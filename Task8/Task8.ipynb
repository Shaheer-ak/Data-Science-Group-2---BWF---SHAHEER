```python
products = {}  
product_order = []  
categories = set()  

# add a product
def add_product(product_id, name, category, stock):
    if product_id in products:
        print(f"Product ID {product_id} already exists.")
    else:
        product_info = (name, category) 
        products[product_id] = {'info': product_info, 'stock': stock}
        product_order.append(product_id)
        categories.add(category)
        print(f"Product {name} added successfully.")

# update stock
def update_stock(product_id, quantity):
    if product_id in products:
        products[product_id]['stock'] += quantity
        print(f"Updated stock for product ID {product_id}. New stock: {products[product_id]['stock']}")
    else:
        print(f"Product ID {product_id} not found.")

# display all products
def display_products():
    print("\nAll Products:")
    for product_id in product_order:
        product = products[product_id]
        info = product['info']
        print(f"ID: {product_id}, Name: {info[0]}, Category: {info[1]}, Stock: {product['stock']}")

# display products by category
def display_by_category(category):
    print(f"\nProducts in Category '{category}':")
    found = False
    for product_id in product_order:
        product = products[product_id]
        info = product['info']
        if info[1] == category:
            found = True
            print(f"ID: {product_id}, Name: {info[0]}, Stock: {product['stock']}")
    if not found:
        print("No products found in this category.")

# Adding products
add_product("ID001", "Product1", "Category1", 50)
add_product("ID002", "Product2", "Category2", 30)
add_product("ID003", "Product3", "Category1", 20)
add_product("ID004", "Product4", "Category3", 10)

display_products()

update_stock("ID001", 20)
update_stock("ID003", -5)

display_products()

display_by_category("Category1")
display_by_category("Category2")
display_by_category("Category3")
display_by_category("Category4")
```
